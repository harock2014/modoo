# 인구 구조를 다양한 형태로 시각화

```python
import csv

f=open('age.csv')
data=csv.reader(f)
result=[]
location = input('인구 구조가 알고 싶은 지역의 이름(읍면동)을 입력해 주세요 :')

for row in data:
    if location in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(location+' 지역의 인구구조')
plt.plot(result)
plt.show()
```

```python
# 두개의 지역을 입력 받아 비교 그래프 생성 
import csv

f=open('age.csv')
data=csv.reader(f)
result1=[]
result2=[]

location1 = input('인구 구조가 알고 싶은 지역의 이름(읍면동)을 입력해 주세요 :')
location2 = input('비교하고 싶은 지역의 이름(읍면동)을 입력해 주세요 :')

for row in data:
    if location1 in row[0]:
        for i in row[3:]:
            result1.append(int(i.replace(',','')))
    if location2 in row[0]:
        for i in row[3:]:
            result2.append(int(i.replace(',','')))

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(location+' 지역과 '+location2+'지역 의 인구구조 비교')
plt.plot(result1, 'r', label=location1)
plt.plot(result2, 'b', label=location2)
plt.legend()
plt.show()
```

## 막대그래프 그리기

bar(막대를 표시할 위치, 막대의 높이)

```python
# 두개의 지역을 입력 받아 하나는 바차트 하나는 plot 차트로 비교 그래프 생성 
import csv

f=open('age.csv')
data=csv.reader(f)
result1=[]
result2=[]

location1 = input('인구 구조가 알고 싶은 지역의 이름(읍면동)을 입력해 주세요 :')
location2 = input('비교하고 싶은 지역의 이름(읍면동)을 입력해 주세요 :')

for row in data:
    if location1 in row[0]:
        for i in row[3:]:
            result1.append(int(i.replace(',','')))
    if location2 in row[0]:
        for i in row[3:]:
            result2.append(int(i.replace(',','')))

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(location+' 지역과 '+location2+'지역 의 인구구조 비교')
plt.bar(range(101), result1, label=location1)
plt.plot(result2, 'b', label=location2)
plt.legend()
plt.show()
```

## 항아리 모양 그래프 그리기

```python
import csv
f=open('gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
            
f.reverse()

import matplotlib.pyplot as plt

plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv
f=open('gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3]))
            f.append(-int(row[-(i+1)]))
f.reverse()

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.title('신드림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

## 지역명을 입력 받아 성별 인구 분포 그래프 생성
