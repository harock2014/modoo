```python
#성범죄자 알림이 지역별 통계 openApi URL 파싱 후 판다스 데이터프레임에 저장

import pandas as pd
# from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET
# import sys
import urllib.request, re
import requests

url='http://116.67.77.182/openapi/SOCitysStats/'

response = requests.get(url)
html = response.text

print(html)

# soup = BeautifulSoup(html, 'html.parser')

# result = re.findall(r'<Result>.+?</Result>', html, re.DOTALL)

data = re.findall(r'<City>.+?</City>', html, re.DOTALL)

print(type(data))

cnt=1

rows=[]

for i in data:
    cityname = re.search(r'<city-name>(.+)</city-name>', i)
    citycount = re.search(r'<city-count>(.+)</city-count>', i)
    print(cnt, cityname, citycount)
    rows.append({
               "cityname":cityname.group(1),
               "citycount":citycount.group(1)})
    cnt = cnt +1

# convert XML data to pandas DataFrame
columns = ["cityname","citycount"]
df = pd.DataFrame(rows,columns = columns )
df
```
