```python
import csv

f=open('age.csv')
data = csv.reader(f)



for row in data:
    if '신도림동' in row[0]:
        cnt = 0
        for i in row[3:]:
            if cnt < 100:
                print(cnt,'세\t\t', i,'명')
            elif cnt >= 100:
                print(cnt,'세이상\t', i,'명')
            cnt = cnt+1
```

특정동 연령별 인구수 데이터 리스트 생성

```python
import csv
f=open('age.csv')
data = csv.reader(f)

result=[]

for row in data:
    if '신도림동' in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)
```

특정동 연령별 인구수 시각화

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data = csv.reader(f)

result=[]

for row in data:
    if '신도림동' in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
