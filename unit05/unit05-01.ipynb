## 데이터리스트에 저장하기

최고 기온을 날짜순으로 저장하는 result 리스트 생성

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)

next(data)
result=[]   # 리스트 자료형

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

print(result)

print(len(result))
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)

next(data)
result=[]   # 리스트 자료형

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.figure(figsize=(10,2))
plt.plot(result, 'r')
plt.show()
```

8월의 최고 기온 데이터만 추출하여 그래프 생성

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)

next(data)
result=[]   # 리스트 자료형

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=='08':
            result.append(float(row[-1]))

plt.figure(figsize=(10,2))
plt.plot(result, 'r')
plt.show()
```

특정일 (2월 14일) 기준 기온 그래프 생성

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)

next(data)
result=[]   # 리스트 자료형

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=='02' and row[0].split('-')[2]=='14':
            result.append(float(row[-1]))

plt.figure(figsize=(10,2))
plt.plot(result, 'r')
plt.show()
```

특정일의 최고 최저 기온 그래프

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)

next(data)
high=[]   # 최고기온 리스트 자료형
low=[] # 최저기온 리스트 자료형

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=='02' and row[0].split('-')[2]=='14':
            high.append(float(row[-1]))
            low.append(float(row[-2]))


plt.plot(high,  'r', label='high')
plt.plot(low, 'b', label='low')
plt.legend()
plt.rc('font', family='Malgun Gothic')   # 한글 글꼴 지정
plt.rcParams['axes.unicode_minus']=False  # 세로축 마이너스 표시 정상화
plt.title('특정일 최고 최저 기온 변화 그래프')
plt.show()
```
